```python
def outstanding_debt(df):
    oustanding_debt_=df['Outstanding_Debt']
    oustanding_debt_list=[]
    
    for i in oustanding_debt_:
        try:
            num=int(i)
            oustanding_debt_list.append(num)
        except:
            string=str(i)
            try:
                string=string.replace('_','0')
            except:
                string=string.replace('-','0')
            float_=float(string)
            num_=np.round(float_)
            num_=int(float_)
            oustanding_debt_list.append(num_)
    return oustanding_debt_list
```

```python
train_df['Outstanding_Debt']=oustanding_debt_list
```

```python
def credit_limit(df):
    credit_limit=df['Changed_Credit_Limit']
    credit_limit_list=[]
    
    for i in credit_limit:
        try:
            num=int(i)
            credit_limit_list.append(num)
        except:
            string=str(i)
            try:
                string=string.replace('_','0')
            except:
                string=string.replace('-','0')
            float_=float(string)
            num_=int(float_)
            credit_limit_list.append(num_)
    return credit_limit_list
```

```python
def emi_per_month(df):
    emi=df['Total_EMI_per_month']
    emi_list=[]
    
    for i in emi:
        try:
            num=int(i)
            emi_list.append(num)
        except:
            string=str(i)
            try:
                string=string.replace('_','')
            except:
                string=string.replace('-','')
            float_=float(string)
            num_=int(float_)
            emi_list.append(num_)
    return emi_list
```

```python
# train_df['Annual_Income']=train_df['Annual_Income'].map(lambda x: int(float(x)))
def annual_income(df):
    anual_income=df['Annual_Income']
    anual_income_list=[]
    
    for i in anual_income:
        try:
            num=int(i)
            anual_income_list.append(num)
        except:
            string=str(i)
            string=string.replace('_','')
            float_=float(string)
            num_=int(float_)
            anual_income_list.append(num_)
    return anual_income_list
```

```python
def cur(df):
    df['Credit_Utilization_Ratio'].convert_dtypes('Float64','int64')
    cur=df['Credit_Utilization_Ratio']
    cur_list=[]
    for i in cur:
        num=int(i)
        cur_list.append(num)
    return cur_list
```

```python
train_df['Credit_Utilization_Ratio']=cur_list
```

```python
def num_delay_payment(df):
    num_delay_payment=df['Num_of_Delayed_Payment']
    num_delay_payment_list=[]
    for i in num_delay_payment:
        try:
            num=int(i)
        except:
            string=str(i)
            string=string.replace('_',' ')
            # string=i.replace('-','')
            num=int(string)
        num_delay_payment_list.append(num)
    return num_delay_payment_list
```
